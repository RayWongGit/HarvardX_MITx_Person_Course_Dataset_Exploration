## Investigation Overview

In this project, we try to explore th pattern of MOOCs certifications and what factors affect certificaton. The variable 'grade' and 'certified' are the main variables of interest, factors we would like to explore in this project including students' demographic features and the students activities.

## Dataset Overview

The dataset we are using, is named 'HMXPC13_DI_v2_5-14-14.csv', which is de-identified data from the first year (Academic Year 2013: Fall 2012, Spring 2013, and Summer 2013) of MITx and HarvardX courses on the edX platform along with related documentation.

These data are aggregate records, and each record represents one individual's activity in one edX course.  

The data description can be found in file 'Person+Course+Documentation.pdf', for better understanding of each variable, I also put the description of each variable into a spreadsheet(named 'HMX Data description.xlsx') with some comments.

The dataset and documents are available on link below:

[dataverse.harvard.edu](https://dataverse.harvard.edu/dataset.xhtml?persistentId=doi:10.7910/DVN/26147)

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
hmx = pd.read_csv('hmx_eda.csv')
hmx.head()
```

## Proportion of Certified VS Uncertified

We see that the proportion is greatly unbalanced. Around 97% students did not earn a certificate.

```python
plt.rcParams.update({'font.size': 16})
plt.figure(figsize=(8, 6))
sb.countplot(data=hmx, x='certified')
n_records = hmx.shape[0]
tick_props = np.arange(0, 1+0.1, 0.1)
tick_names = ['{:.1f}'.format(x) for x in tick_props]
plt.yticks(tick_props*n_records, tick_names)
labels = ['Uncertified', 'Certified']
plt.xticks(np.arange(2), labels=labels)
plt.title('Proportion of Certified VS Uncertified')
plt.xlabel('Whether or not Certified')
plt.ylabel('Proportion');
```

## Does age affect the performance?

Although some younger students also get low grades, in general, younger student(20-30 years old) are more likely to get higher grades, which is to say, reach better performance than elder students

```python
plt.figure(figsize=(10, 8))
plt.scatter(data=hmx, x='age', y='grade',alpha=0.05)
plt.title('Relationship between grade and age')
plt.xlabel('Age')
plt.ylabel('Grade');
```

## Distribution of Age by Education level and Gender

The average age increase as the level of highest education increase, but there is no significant diffrence between male and female with the same education background.


```python
plt.figure(figsize=(12, 8))
order = ['Less than Secondary','Secondary', "Bachelor's", "Master's", 'Doctorate']
sb.violinplot(data=hmx, x='LoE_DI', y='age', hue='gender', order=order)
plt.title('Distribution of Age by Education level and Gender')
plt.xlabel('Highest level of education')
plt.ylabel('Age')
plt.legend(loc='upper left');
```
